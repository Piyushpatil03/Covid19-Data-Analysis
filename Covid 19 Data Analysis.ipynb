```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Downloads/covid19_Confirmed_dataset.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data.drop(["Lat","Long"],axis=1, inplace=True)
```

```python
data.head()
```

```python
# There are more Null values in Province/State.
# We have to aggregate data by Country.
df = data.groupby("Country/Region").sum()
```

```python
df.head()
```

# Visualizing the data

```python
plt.figure(figsize=(10,4))
df.loc["China"].plot()
df.loc["Italy"].plot()
df.loc["India"].plot()
plt.legend()
```

```python
## Calculating first derivative of curve
df.loc["China"].diff().plot()
```

```python
df.loc["China"].diff().max()
```

```python
## find maximum infection rate for all of the countries.

countries = list(df.index)

max_rates = []
for c in countries :
    max_rates.append(df.loc[c].diff().max())
    
max_rates
```

```python
df["max_rates"] = max_rates
```

```python
df.head()
```

```python
corona_data = pd.DataFrame(df["max_rates"])
```

```python
corona_data.head()
```

```python
happiness_data = pd.read_csv("Downloads/worldwide_happiness_report.csv")
```

```python
happiness_data.head()
```

```python
useless_columns = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_data.drop(useless_columns,axis=1,inplace=True)
```

```python
happiness_data.head()
```

```python
happiness_data.shape
```

```python
corona_data.shape
```

```python
happiness_data.set_index("Country or region",inplace=True)
```

# Merge both data

```python
final = corona_data.join(happiness_data,how="inner")
final.head()
```
